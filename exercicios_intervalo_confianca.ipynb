# 1. Sampling Distribution & Confidence Interval

## 📚 Teoria

A distribuição amostral da média tem propriedades fundamentais:

- Média: μ
- Erro padrão: 
    SE = σ / √n

Pelo Teorema Central do Limite:
A média amostral segue aproximadamente uma distribuição normal.

### Intervalo de Confiança:

CI = x̄ ± Z * (σ / √n)

Onde:
- x̄ = média da amostra
- Z = valor crítico (ex: 1.96 para 95%)

## 🎯 Exercício

Simule uma população com distribuição normal (μ=50, σ=10).

1. Extraia 1000 amostras de tamanho 30
2. Calcule a média de cada amostra
3. Plote a distribuição das médias
4. Compare com μ real


```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

population = np.random.normal(50, 10, 100000)

sample_means = []

for _ in range(1000):
    sample = np.random.choice(population, 30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30)
plt.axvline(np.mean(population), color='red')
plt.title('Sampling Distribution of Mean')
plt.show()

print("Mean of sample means:", np.mean(sample_means))
```

# 2. Construção de Intervalo de Confiança

## 📚 Teoria

Quando σ é desconhecido, usamos:

CI = x̄ ± t * (s / √n)

Onde:
- s = desvio padrão amostral
- t = valor crítico da t-student

## 🎯 Exercício

Crie uma função que calcula o intervalo de confiança para 95%.

Entrada: array de dados
Saída: (lower, upper)

```
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    se = stats.sem(data)
    h = se * stats.t.ppf((1 + confidence) / 2, n - 1)
    return mean - h, mean + h

sample = np.random.normal(50, 10, 50)

confidence_interval(sample)
```

# 3. Difference in Means (A/B Test Core)

## 📚 Teoria

Diferença de médias:

Δ = x̄₁ - x̄₂

Erro padrão da diferença:

SE = √(s₁²/n₁ + s₂²/n₂)

CI:

Δ ± Z * SE

## 🎯 Exercício

Simule dois grupos:
- Grupo A: μ=50
- Grupo B: μ=55

1. Calcule Δ
2. Construa CI da diferença
3. Interprete se há diferença significativa


```
group_A = np.random.normal(50, 10, 1000)
group_B = np.random.normal(55, 10, 1000)

mean_A = np.mean(group_A)
mean_B = np.mean(group_B)

diff = mean_B - mean_A

se = np.sqrt(np.var(group_A)/len(group_A) + np.var(group_B)/len(group_B))

ci_lower = diff - 1.96 * se
ci_upper = diff + 1.96 * se

diff, ci_lower, ci_upper
```
