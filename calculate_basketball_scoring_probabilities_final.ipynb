## Section 1 — Formula

**Empirical (frequentist) probability** — estimated directly from data:

$$P(A) = \frac{\text{count}(A)}{N}$$

where $N$ is the total number of observations.

Applied here:

$$P(\text{make}) = \frac{\text{count}(\text{result}=\text{True})}{N}$$

$$P(\text{miss}) = 1 - P(\text{make})$$

$$P(\text{3-pointer}) = \frac{\text{count}(\text{shot\_type}=3)}{N}$$

$$P(\text{2-pointer}) = \frac{\text{count}(\text{shot\_type}=2)}{N}$$

## Section 2 — Formula

**Binomial distribution** — probability of exactly $k$ successes in $n$ independent trials:

$$P(X = k) = \binom{n}{k} \, p^k \, (1-p)^{n-k}$$

| Symbol | Meaning |
|--------|----------------------------------|
| $n$    | number of trials (shots)         |
| $k$    | desired number of successes      |
| $p$    | probability of success per trial |

**Assumptions:** shots are independent and $p$ is constant across shots.

---

**Applied numerically:**

**Q1** — P(exactly 3 makes in 4 shots), with $p = 0.4902$:

$$P(X=3) = \binom{4}{3} \times 0.4902^3 \times 0.5098^1 = 4 \times 0.1177 \times 0.5098 \approx 0.2402$$

**Q2** — P(exactly 4 three-pointers in 5 shots), with $p = 0.5481$:

$$P(X=4) = \binom{5}{4} \times 0.5481^4 \times 0.4519^1 = 5 \times 0.0902 \times 0.4519 \approx 0.2039$$

## Section 3 — Formula

**Conditional probability** — probability of $A$ given that $B$ already occurred:

$$P(A \mid B) = \frac{P(A \cap B)}{P(B)} = \frac{\text{count}(A \text{ and } B)}{\text{count}(B)}$$

The joint probability is calculated as:

$$P(A \cap B) = \frac{\text{count}(A \text{ and } B)}{N}$$

**Key rule:** the notation $P(A \mid B)$ means:
- **Left of** $\mid$ → what you want to discover
- **Right of** $\mid$ → evidence you already have

Applied in this section:

$$P(\text{make} \mid \text{three}) = \frac{P(\text{make} \cap \text{three})}{P(\text{three})}$$

$$P(\text{lead} \mid \text{three}) = \frac{P(\text{lead} \cap \text{three})}{P(\text{three})}$$

$$P(\text{make} \mid \text{two}) = \frac{P(\text{make} \cap \text{two})}{P(\text{two})}$$

$$P(\text{lead} \mid \text{two}) = \frac{P(\text{lead} \cap \text{two})}{P(\text{two})}$$

## Section 4 — Formula

**Bayes' Theorem** — inverts the conditional: you have $P(B \mid A)$ but want $P(A \mid B)$:

$$P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}$$

| Term | Name | Meaning |
|------|------|---------|
| $P(A)$ | **Prior** | Belief about $A$ before seeing evidence |
| $P(B \mid A)$ | **Likelihood** | Probability of evidence given cause |
| $P(B)$ | **Evidence** | Total probability of observing $B$ |
| $P(A \mid B)$ | **Posterior** | Updated belief after evidence |

Applied in this section:

$$P(\text{three} \mid \text{make}) = \frac{P(\text{make} \mid \text{three}) \cdot P(\text{three})}{P(\text{make})}$$

$$P(\text{two} \mid \text{make}) = \frac{P(\text{make} \mid \text{two}) \cdot P(\text{two})}{P(\text{make})}$$

> **Why Bayes?** In Section 3 you had the evidence (shot type) and wanted the outcome.
> Here the direction is **reversed**: you know the outcome (make) and want to infer the shot type.

## Section 4 — Conditional Probabilities (Past)

### Questions to Answer

If Steph **just made a shot**:
1. What is the probability that it was a **three-pointer**?
2. What is the probability that it was a **two-pointer**?

> **Note:** The project says you *could* answer these by direct counting,
> but you **must** use Bayes' Theorem and show your steps.

---

````python
# Section 4: Conditional Probabilities (Past) — Bayes' Theorem
# P(three | make) = P(make | three) * P(three) / P(make)
# P(two   | make) = P(make | two)   * P(two)   / P(make)

# Step 1: P(make | two) — likelihood needed for Ex2 (not yet calculated)
count_make_and_two = df[df['result'] & (df['shot_type'] == 2)].shape[0]
p_make_and_two = count_make_and_two / total_shots
p_make_given_two = p_make_and_two / p_two

print(f"Count(make AND two):  {count_make_and_two}")
print(f"P(make | two):        {p_make_given_two:.4f}")
print(f"P(make | three):      {p_make_given_three:.4f}  (from Section 3)")
````

````python
# Step 2: Apply Bayes' Theorem
# Ex1: P(three | make) = P(make | three) * P(three) / P(make)
p_three_given_make = (p_make_given_three * p_three) / p_make

# Ex2: P(two | make) = P(make | two) * P(two) / P(make)
p_two_given_make = (p_make_given_two * p_two) / p_make

print(f"P(three | make): {p_three_given_make:.4f}")
print(f"P(two   | make): {p_two_given_make:.4f}")
````

````python
print("=" * 60)
print(f"{'SECTION 4 — CONDITIONAL PROBABILITIES (PAST)':^60}")
print("=" * 60)
print(f"\n{'Bayes Theorem components':}")
print(f"  P(make | three) = {p_make_given_three:.4f}  (likelihood)")
print(f"  P(make | two)   = {p_make_given_two:.4f}  (likelihood)")
print(f"  P(three)        = {p_three:.4f}  (prior)")
print(f"  P(two)          = {p_two:.4f}  (prior)")
print(f"  P(make)         = {p_make:.4f}  (evidence)")
print()
print(f"{'Question':<40} {'Probability':>10}")
print("-" * 55)
print(f"{'P(three | make)':<40} {p_three_given_make:>10.4f}")
print(f"{'P(two   | make)':<40} {p_two_given_make:>10.4f}")
print("=" * 60)
print(f"\nSanity check — P(three|make) + P(two|make) = {p_three_given_make + p_two_given_make:.4f} (must be 1.0)")
````

### Ex3 — Retrospective: P(lead | make)

> **Retrospective probability:** we already know the outcome (Steph *made* the shot) 
and are inferring the context (was the team leading?) — reasoning backward from result to circumstance.

---

**Terms needed:**

| Term | Name | Value |
|------|------|-------|
| $P(\text{lead})$ | Prior | proportion of shots taken while leading |
| $P(\text{make} \mid \text{lead})$ | Likelihood | make rate when team was leading |
| $P(\text{make})$ | Evidence | overall make rate |

**Via Bayes' Theorem:**

$$P(\text{lead} \mid \text{make}) = \frac{P(\text{make} \mid \text{lead}) \cdot P(\text{lead})}{P(\text{make})}$$

**Equivalently, via direct joint probability** (both give the same result):

$$P(\text{lead} \mid \text{make}) = \frac{P(\text{make} \cap \text{lead})}{P(\text{make})} = \frac{\text{count}(\text{make AND lead})}{\text{count}(\text{make})}$$

````python
# Ex3: P(lead | make) — retrospective conditional probability
# Approach A — Bayes: P(lead|make) = P(make|lead) * P(lead) / P(make)
# Approach B — Direct joint: P(lead|make) = P(make ∩ lead) / P(make)

# Step 1: compute components
count_lead          = df['lead'].sum()
count_make_and_lead = df[df['result'] & df['lead']].shape[0]

p_lead              = count_lead / total_shots                    # prior
p_make_given_lead   = count_make_and_lead / count_lead            # likelihood
p_make_and_lead     = count_make_and_lead / total_shots           # joint

# Step 2: apply both approaches
p_lead_given_make_bayes  = (p_make_given_lead * p_lead) / p_make  # Bayes
p_lead_given_make_direct = p_make_and_lead / p_make               # direct joint

print("Bayes components:")
print(f"  P(lead):           {p_lead:.4f}  (prior)")
print(f"  P(make | lead):    {p_make_given_lead:.4f}  (likelihood)")
print(f"  P(make):           {p_make:.4f}  (evidence)")
print(f"  P(make ∩ lead):    {p_make_and_lead:.4f}  (joint)")
print()
print(f"P(lead | make) via Bayes:       {p_lead_given_make_bayes:.4f}")
print(f"P(lead | make) via direct joint:{p_lead_given_make_direct:.4f}")
print(f"Sanity check — both match: {abs(p_lead_given_make_bayes - p_lead_given_make_direct) < 1e-10}")
````

````python
# P(make | two) = P(make ∩ two) / P(two)
count_make_and_two = df[df['result'] & (df['shot_type'] == 2)].shape[0]
p_make_and_two = count_make_and_two / total_shots
p_make_given_two = p_make_and_two / p_two

print(f"Count(make AND two): {count_make_and_two}")
print(f"P(make ∩ two):       {p_make_and_two:.4f}")
print(f"P(two):              {p_two:.4f}")
print(f"P(make | two):       {p_make_given_two:.4f}")
````

````python
# P(lead | two) = P(lead ∩ two) / P(two)
count_lead_and_two = df[df['lead'] & (df['shot_type'] == 2)].shape[0]
p_lead_and_two = count_lead_and_two / total_shots
p_lead_given_two = p_lead_and_two / p_two

print(f"Count(lead AND two): {count_lead_and_two}")
print(f"P(lead ∩ two):       {p_lead_and_two:.4f}")
print(f"P(two):              {p_two:.4f}")
print(f"P(lead | two):       {p_lead_given_two:.4f}")
````
